### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Construct Data**

Pada dataset ini, saya memutuskan untuk menambahkan beberapa kolom dengan tujuan meningkatkan kualitas dari dataset ini. Kolom tersebut diantaranya:

1. **customer_age_category**

```python
def customer_age_category(customer_age):
    if customer_age < 18:
        return 'Underage'
    elif 18 <= customer_age <= 35:
        return 'Young Adult'
    elif 36 <= customer_age <= 55:
        return 'Adult'
    else:
        return 'Senior'

df['customer_age_category'] = df['customer_age'].apply(customer_age_category)
df
```

Kolom customer_age_category merupakan hasil pengelompokan dari kolom customer_age menjadi kategori “Underage” untuk usia < 18 tahun, “Young Adult” untuk usia 18–35 tahun, “Adult” untuk usia 36–55 tahun, dan “Senior” untuk usia > 55 tahun. Penambahan kolom ini bertujuan untuk menyederhanakan dan mengelompokkan data usia pelanggan ke dalam kategori yang lebih mudah dipahami. Kolom ini berguna untuk meningkatkan kualitas dataset karena mempermudah analisis perilaku pelanggan berdasarkan kelompok usia, serta memudahkan proses analisis dan visualisasi data tanpa menghilangkan makna dari variabel customer_age.

2. **order_value_category**

```python
def order_value_category(usd):
    if usd < 250:
        return 'Low Value'
    elif 250 <= usd < 1000:
        return 'Medium Value'
    else:
        return 'High Value'

df['order_value_category'] = df['order_value_usd'].apply(order_value_category)
df
```

Kolom order_value_usd_category merupakan hasil pengelompokan dari kolom order_value_usd menjadi kategori “Low Value” untuk nilai pesanan rendah, “Medium Value” untuk nilai pesanan sedang, dan “High Value” untuk nilai pesanan tinggi. Penambahan kolom ini bertujuan untuk menyederhanakan dan mengelompokkan data nilai transaksi dalam USD agar lebih mudah diinterpretasikan. Kolom ini berguna untuk meningkatkan kualitas dataset karena membuat dataset lebih informatif, mempermudah analisis perilaku pelanggan berdasarkan nilai transaksi, serta memudahkan analisis dan visualisasi data tanpa menghilangkan makna dari variabel order_value_usd.

3. **customer_rating_category**

```python
def customer_rating_category(rating):
    if rating < 3:
        return 'Low'
    elif 3 <= rating < 4:
        return 'Medium'
    else:
        return 'High'

df['customer_rating_category'] = df['customer_rating'].apply(customer_rating_category)
df
```

Kolom customer_rating_category merupakan hasil pengelompokan dari kolom customer_rating menjadi kategori "Low" untuk skala 1-2, "Medium" untuk skala 3, dan "High" untuk skala 4-5. Penambahan kolom ini bertujuan untuk menyederhanakan dan mengelompokan data dari penilaian pelanggan. Kolom ini berguna untuk meningkatkan kualitas dataset karena membuat dataset lebih informatif, serta mempermudah analisis dan visualisasi data tanpa perlu menghilangkan makna dari variabel customer_rating.

## **Line Chart**

```python
# Ambil tahun dari order_date
df['year'] = df['order_date'].dt.year

# Total nilai pesanan per tahun
VARIABEL = df.groupby('year')['order_value_usd'].mean()

plt.figure(figsize=(10,6))
plt.plot(VARIABEL.index, VARIABEL.values, marker='o')

plt.xlabel('Tahun')
plt.ylabel('Rata-rata Nilai Pesanan (USD)')
plt.title('Tren Penjualan dari Tahun ke Tahun')

plt.xticks(VARIABEL.index)
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Insight**

Berdasarkan line chart rata-rata nilai pesanan per tahun, terlihat bahwa nilai transaksi pelanggan mengalami perubahan dari tahun ke tahun. Rata-rata nilai pesanan mencapai nilai yang cukup tinggi pada tahun 2023 dan 2033, sedangkan penurunan terlihat pada tahun 2024 serta sekitar tahun 2028-2030 sebelum kembali meningkat pada tahun-tahun berikutnya. Meskipun terjadi perubahan, rata-rata nilai pesanan pelanggan masih berada pada kisaran yang relatif stabil, yaitu sekitar 730 hingga 780 USD.

**Action**

Berdasarkan *insight* tersebut, perusahaan *e-commerce* dapat mempelajari faktor dan straegi penjualan pada tahun dengan rata rata nilai pesanan yang tinggi untuk diterapkan pada tahun-tahun berikutnya. Selain itu, pada tahun yang mengalami penurunan rata-rata nilai pesanan, perusahaan dapat melakukan tindakan dengan mendorong penawaran diskon dan strategi rekomendasi produk agar pelanggan terdorong melakukan pembelian dengan nilai yang lebih besar.

## **BoxPlot**

```python
top3payment = df['payment_method'].value_counts().head(3).index

df_top3_payment = df[df['payment_method'].isin(top3payment)]

plt.figure(figsize=(12, 7)) # Ukuran figure
df_top3_payment.boxplot(column='order_value_usd', by='payment_method', ax=plt.gca(), grid=False, patch_artist=True, medianprops=dict(color='red'))
plt.title('Distribusi Nilai Pesanan di Top 3 Metode Pembayaran', fontsize=16) # Judul
plt.xlabel('Metode Pembayaran', fontsize=12) # X Keterangan
plt.ylabel('Nilai Pesanan (USD)', fontsize=12) # Y Keterangan
plt.xticks(rotation=45, ha='right')
plt.suptitle('') # Menghilangkan judul default dari boxplot.by
plt.tight_layout()
plt.show()
```

Median nilai pesanan pada ketiga metode berada di kisaran yang hampir sama, menandakan bahwa pelanggan cenderung melakukan transaksi dengan nilai yang tidak jauh berbeda terlepas dari metode pembayaran yang digunakan. Selain itu, rentang nilai pesanan yang cukup lebar pada masing-masing metode menunjukkan bahwa metode pembayaran tersebut digunakan baik untuk transaksi bernilai rendah maupun tinggi.
