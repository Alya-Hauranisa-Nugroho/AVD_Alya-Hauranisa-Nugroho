### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

**Aktivitas:** Membandingkan total nilai pesanan (*order value*) pada setiap metode pembayaran.

**Tujuan:** Mengidentifikasi metode pembayaran yang memberikan kontribusi nilai transaksi terbesar serta membandingkan performa masing-masing metode pembayaran secara langsung.

**Visualisasi:** Bar Chart (Grafik Batang).

```python
penjualan = df.groupby('payment_method')['order_value_usd'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan.index, y=penjualan.values, palette='viridis', hue=penjualan.index, legend=False)
plt.title('Nilai Pesanan Berdasarkan Metode Pembayaran')
plt.xlabel('Metode Pembayaran')
plt.ylabel('Nilai Pesanan (USD)')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan** : Metode pembayaran Credit Card dan Wallet memiliki total nilai pesanan tertinggi, menunjukkan kecenderungan pelanggan melakukan transaksi bernilai besar dengan pembayaran digital. Sebaliknya, Cash on Delivery dan Debit Card cenderung digunakan untuk transaksi dengan nilai yang lebih rendah.

## **Composition/Komposisi**

**Aktivitas:** Melihat proporsi nilai pesanan (*order value*) berdasarkan kategori produk.

**Tujuan:** Mengetahui kategori produk mana yang memberikan kontribusi terbesar terhadap total nilai pesanan.

**Visualisasi:** Pie Chart (Grafik Lingkaran)

```python
sales_by_category = df.groupby('product_category')['order_value_usd'].sum().sort_values(ascending=False)
sales_by_category_top3 = sales_by_category.head(6)

plt.figure(figsize=(10, 8))
sales_by_category_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 6 Kategori Produk Teratas Berdasarkan Nilai Pesanan')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

**Kesimpulan:** Proporsi nilai pesanan pada enam kategori produk teratas relatif merata, dengan Fashion sebagai kontributor terbesar, namun selisihnya tipis dibanding kategori lain.

## **Distribution/Distribusi**

**Aktivitas:** Menganalisis penyebaran nilai pesanan pada kolom order_value_usd.

**Tujuan:** Mengetahui apakah mayoritas pesanan bernilai rendah atau tinggi, serta mendeteksi adanya kecondongan distribusi (*skewness*) pada nilai pesanan pelanggan.

**Visualisasi:** Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['order_value_usd'], bins=20, kde=True)
plt.title('Distribusi Nilai Pesanan (USD)')
plt.xlabel('Frekuensi')
plt.ylabel('Frekuensi')
plt.show()
```

**Kesimpulan:** Distribusi nilai pesanan tersebar cukup merata di berbagai rentang harga tanpa kecondongan ekstrem ke nilai rendah atau tinggi. Hal ini menunjukkan variasi nilai transaksi yang beragam dan tidak adanya *outlier* yang dominan, sehingga pola distribusi cenderung mendekati simetris.

## **Relationship/Hubungan**

**Aktivitas:** Menganalisis hubungan antara lama waktu pengiriman (delivery_time_days) dengan penilaian pelanggan (customer_rating).

**Tujuan:** Mengetahui apakah semakin lama waktu pengiriman berpengaruh terhadap penurunan penilaian pelanggan.

**Visualisasi:** Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['delivery_time_days', 'customer_rating']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Lama Pengiriman dan Penilaian Pelanggan')
plt.show()
```

**Kesimpulan:** Nilai korelasi antara lama pengiriman dan penilaian pelanggan sebesar -0,02, yang menunjukkan hubungan sangat lemah dan hampir tidak ada korelasi. Artinya, lama pengiriman tidak berpengaruh signifikan terhadap penilaian pelanggan pada dataset ini.
